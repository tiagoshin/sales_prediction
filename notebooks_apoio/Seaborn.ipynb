```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,kind='kde')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

# categorical

```python
import numpy as np
```

```python
#estimator custom =mean
sns.barplot(x='sex',y='total_bill', data=tips, estimator=np.std)
# 1 categorical and 1 numerical
```

```python
sns.countplot(x='sex', data=tips)
# 1 categorical and 1 numerical
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
#1 categorical and 1 numerical
#hue separar por um categorical (yes or no nesse caso)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',split=True)
#1 cate 1 numericaal
#show distribution of all the points
# split  usado junto com hue
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
#1 cat 1 num
#Mostra distribuio
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
#mesma coisa, porm usa o espao lateral
# por isso no  recomendado pra datasets muito grandes
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',split=True)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips)
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time',row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time',row='smoker')
g.map(plt.scatter, 'total_bill','tip')
```

# Regression plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':5})
#scatter_kws s  o tamanho das bolinhas
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', hue='sex', aspect=0.6, size=8)
```
